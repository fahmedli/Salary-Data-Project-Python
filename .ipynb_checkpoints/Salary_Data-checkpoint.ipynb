# Visualization, Graphs

```python
salary_df=df.groupby('Job Title')['Salary'].mean()
```

```python
salary_df
```

```python
salary_df.reset_index()
```

```python
import matplotlib.pyplot as py
```

```python
salary_df.plot(x="Job Title", y="Salary")
```

```python
salary_df = pd.DataFrame({
    'Job Title': salary_df.index,  
    'Salary': salary_df.values     
})
```

```python
salary_df.sort_values("Salary", ascending=False).head(20).plot(x="Job Title", y="Salary",kind='barh' )
```
